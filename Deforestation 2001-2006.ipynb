```python
deforestation = deforestation.set_index('year')
deforestation
```

```python
deforestation = deforestation.div(10000)
#area now in ha
deforestation
```

```python
deforestation_flipped = deforestation.T
deforestation_flipped
```

```python
first_period = [2001,2002, 2003, 2004,2005]
deforestation_flipped['2001-2005'] = deforestation_flipped[first_period].sum(axis=1)
deforestation_flipped = deforestation_flipped.rename(columns={15:"Total"})
```

```python
deforestation_flipped = deforestation_flipped.reset_index()
```

```python
deforestation_flipped = deforestation_flipped.rename(columns={"index":"Country"})
```
